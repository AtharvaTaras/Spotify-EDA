```python
clean_df = data.drop_duplicates(subset=['track_id'])
clean_df.shape
```

```python
most_popular = clean_df.sort_values('popularity', ascending=False)
most_popular.head(10)[['artist_name', 'track_name', 'genre', 'popularity']]

# Top 10 Most Popolar Songs
```

```python
most_popular.tail(10)[['artist_name', 'track_name', 'genre', 'popularity']]

# 10 Least Popular Songs
```

```python
zero_popularity = clean_df.query('popularity==0', inplace=False)
zero_popularity.shape[0]

# Number of songs with 0 popularity
```

```python
track_len = clean_df.sort_values('duration_ms', ascending=True)
track_len['duration_ms'] = track_len['duration_ms'].apply(lambda x: round(x/1000))
```

```python
track_len.head(10)[['track_name', 'artist_name', 'duration_ms']]

# 10 Shotest Tracks (duration in seconds)
```

```python
track_len.tail(10)[['track_name', 'artist_name', 'duration_ms']]

# 10 Longest Tracks (duration in seconds)
```
